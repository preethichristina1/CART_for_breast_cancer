```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\HP\OneDrive\Desktop\data.csv")
```

```python
df.head(5)
```

```python
X=df.drop(['diagnosis','Unnamed: 32'],axis=1)
```

```python
X.head(5)
```

```python
y=df['diagnosis']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=101)
```

```python
y_train
```

```python
from sklearn import preprocessing
encode=preprocessing.LabelEncoder()
```

```python
l_train=encode.fit_transform(y_train)
l_test=encode.transform(y_test)
```

```python
Y=np.array(l_train).reshape(-1,1)
Y
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier(criterion='entropy',max_depth=6,random_state=101)
```

```python
dtc.fit(X_train,Y)
```

```python
predicted=dtc.predict(X_test)
predicted
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(l_test,predicted)
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(l_test,predicted)
acc
```

```python
from sklearn.metrics import precision_score
precision= precision_score(l_test,predicted)
precision
```
